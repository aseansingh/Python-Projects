```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_titanic = pd.read_csv("titanic.csv")
```

- Consider the following features and draw some conclusions about survival or no survival using any kind of visualization.

    1. age
    2. Pclass
    3. Sex
      
- Fare and age


```python
## Your code here

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=df_titanic, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Survival based on Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='Pclass', hue='Survived')
plt.title('Survival based on Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='Sex', hue='Survived')
plt.title('Survival based on Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Survived')
plt.title('Survival based on Fare and Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
## Your code here

import warnings
warnings.filterwarnings("ignore")

sns.pairplot(df_titanic[["Survived", "Pclass", "Sex", "Age", "Siblings/Spouses Aboard", "Parents/Children Aboard"]],
             hue="Sex")

plt.show()

"""The histograms along the diagonal shows the distribution of passengers by sex for each feature. The overlaid density 
plots for males and females help to compare the two groups. The scatter plots make the diagonal illustrate the relationships 
between pairs of features. The color coding differentiates male and female passengers, allowing for a visual comparison 
on each sex fares across different feature combinations. Color usage in the Survived plot indicates the survival rates 
by sex also shows higher survival rates for females. The Pclass and Age plots colored by sex provide the insights into the 
age distribution of male and female passengers aboard."""
```

```python
The image is a boxplot that shows the distribution of age of passengers on the Titanic, broken down by Pclass (passenger class).
From the graph, it shows that a relationship is between a passenger age and their Pclass.
1. Passengers in Pclass 1 (the wealthiest class) tend to be older than those in Pclass 2 or 3.
2. The median age in Pclass 1 is higher than the medians in the other two classes.
3. There is more variability in the ages of passengers in Pclass 1 than in Pclass 3. 
4. The box for Pclass 1 is taller than the box for Pclass 3.
5. There may be a few outliers in Pclass 1 that are much older than the rest of the passengers. 
6. The top whisker for Pclass 1 extends much farther than the whiskers for the other two classes.
```

```python
# Draw the followinmg graph and comment

class1_data = df_titanic[df_titanic['Pclass'] == 1]['Age']
class2_data = df_titanic[df_titanic['Pclass'] == 2]['Age']
class3_data = df_titanic[df_titanic['Pclass'] == 3]['Age']

plt.hist([class1_data, class2_data, class3_data], bins=30, stacked=True)

plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(['Pclass 1', 'Pclass 2', 'Pclass 3'])
 
plt.show()

""" The below stacked histogram visualizes the age distribution of passengers in 
different passenger classes on the Titanic using a stacked histogram. 
It first extracts age data for passengers in each passenger class, 
then plots the histogram with age on the x-axis and count on the y-axis. 
Each passenger class is represented by a different color, 
and the legend indicates which color corresponds to each class."""
```
