Q1 what is visualization. what is the purpose?  
Ans Visualisation is the process of creating graphical representation of images to communicate information or ideas. It involves creating images, diagrams, animation to communicate a message. These can be formation of mental visual images, representation of data using charts, graphs, maps.
The purpose of visualization is :-
1. Visualisation helps in data cleaning through identifying errors, incosistency and outliers in the data.
2. The data structure in visualisation helps in understanding the underlying patterns and structure of the data.
3. Visualisation helps in detecting outliers and unusual groups while identifying data points that deviate significantly.
4. Visualisation helps in identifying trends and clusters in data over time and groupings of related data points.
5. Visualisation spots local patterns which highlight patterns in a specific subset of data.
6. Visualisation evaluates model output through performance of statistical model.
7. Visualisation presents results of data analysis in accessible and understandable form.
8. Visualisation presents data in visual format which helps in decision making.

Q11
Create the following graph using the following properties

We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or axis('tight') for automatically getting "tightly fitted" axes ranges:

```python

x = np.linspace(0, 5, 11)
y1 = x ** 2
y2 = x ** 3

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].plot(x, y1, label='x**2')
axs[0].plot(x, y2, label='x**3')
axs[0].set_title('default axes ranges')
axs[0].legend()

axs[1].plot(x, y1, label='x**2')
axs[1].plot(x, y2, label='x**3')
axs[1].axis('tight')
axs[1].set_title('tight axes')
axs[1].legend()

axs[2].plot(x, y1, label='x**2')
axs[2].plot(x, y2, label='x**3')
axs[2].set_xlim([2, 5])
axs[2].set_ylim([0, 60])
axs[2].set_title('custom axes range')
axs[2].legend()

fig.suptitle('multiple plots')

plt.show()
```

Q12 Create the following using data in x and y 

```python

x = np.linspace(0, 5, 100)
y1 = x**2
y2 = x

fig, ax1 = plt.subplots()

plt.title("Axes 2 Title")
ax1.plot(x, y1, 'b-')
ax1.set_xlabel('X_label_axes2')
ax1.set_ylabel('Y_label_axes2')
ax1.tick_params('x')

left, bottom, width, height = [0.2, 0.5, 0.4, 0.3]
ax2 = fig.add_axes([left, bottom, width, height])
plt.title("Axes 2 Title")
ax2.plot(x**2, y2, 'r-')
ax2.set_xlabel('X_label_axes2')
ax2.set_ylabel('Y_label_axes2')
ax2.tick_params('x')
ax2.set_xticks([0, 10, 20])

plt.show()
```
